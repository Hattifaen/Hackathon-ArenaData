```python
import pandas as pd
from pathlib import Path

# Путь к файлу
file_path = Path('resultik.csv')

# Чтение и фильтрация без вывода
if file_path.exists():
    df = pd.read_csv(file_path)
    # if 'total_hits' in df.columns:
    #     df = df[df['total_hits'] != 0].reset_index(drop=True)
    # else:
    #     df = pd.DataFrame()  # или оставить как есть, если столбца нет
else:
    df = pd.DataFrame()  # пустой DataFrame, если файл не найден
```

```python
import pandas as pd
import os

# Убедимся, что total_hits — числовой тип
df['total_hits'] = pd.to_numeric(df['total_hits'], errors='coerce')

# Фильтруем строки, где total_hits не равно 0
df_nonzero = df[df['total_hits'] != 0].copy()

# Создаём колонку name, извлекая имя файла из пути
df_nonzero['name'] = df_nonzero['path'].apply(os.path.basename)

# Оставляем только колонку 'name'
result = df_nonzero[['name']].reset_index(drop=True)
```
